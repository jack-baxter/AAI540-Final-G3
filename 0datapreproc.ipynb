```python
import pandas as pd
import numpy as np 
```

```python
aimodelpoll = pd.read_csv('aimodelpoll.csv')
lifearch = pd.read_csv('lifearchitectmodels.csv')
llmach = pd.read_csv('llmachievements.csv')
llmprice = pd.read_csv('llmpricingdata.csv')
llmleader = pd.read_csv('openllmleader.csv')
modelview = pd.read_csv('overviewaimodels.csv')
```

```python
aimodelpoll.head(5)
```

```python
aimodelpoll.info()
```

```python
for col in aimodelpoll.columns[2:]:
    aimodelpoll[col] = (
        aimodelpoll[col]
        .astype(str)                           
        .str.replace(r'[\s%,]', '', regex=True) 
        .str.replace(',', '.', regex=False)   
        .replace(['', 'nan', 'NaN', '-'], np.nan)
        .pipe(pd.to_numeric, errors='coerce')
        / 100
    )
aimodelpoll.iloc[:, 2:] = aimodelpoll.iloc[:, 2:].round(3)
```

```python
aimodelpoll.head(5)
```

```python
aimodelpoll.info()
```

```python
original_file = 'aimodelpoll.csv'
aimodelpoll.to_csv(original_file, index=False)
```

```python
llmach.head(5)
```

```python
llmach.info()
```

```python
for col in ['Result', 'Human result']:
    llmach[col] = (
        llmach[col]
        .astype(str)                                 
        .str.replace(r'[\s%,]', '', regex=True)      
        .str.replace(',', '.', regex=False)         
        .replace(['', 'nan', 'NaN', '-', '--'], np.nan)
        .pipe(pd.to_numeric, errors='coerce')        
    )
llmach['Result'] = llmach['Result'].round(2)
llmach['Human result'] = llmach['Human result'].round(2)
```

```python
llmach.head(5)
```

```python
original_file = 'llmachievements.csv'
llmach.to_csv(original_file, index=False)
```
