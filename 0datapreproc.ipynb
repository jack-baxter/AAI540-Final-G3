```python
import pandas as pd
import numpy as np 
```

```python
aimodelpoll = pd.read_csv('aimodelpoll.csv')
lifearch = pd.read_csv('lifearchitectmodels.csv')
llmach = pd.read_csv('llmachievements.csv')
llmprice = pd.read_csv('llmpricingdata.csv')
llmleader = pd.read_csv('openllmleader.csv')
modelview = pd.read_csv('overviewaimodels.csv')
```

```python
aimodelpoll.head(5)
```

```python
aimodelpoll.info()
```

```python
for col in aimodelpoll.columns[2:]:
    aimodelpoll[col] = (
        aimodelpoll[col]
        .astype(str)                           
        .str.replace(r'[\s%,]', '', regex=True) 
        .str.replace(',', '.', regex=False)   
        .replace(['', 'nan', 'NaN', '-'], np.nan)
        .pipe(pd.to_numeric, errors='coerce')
        / 100
    )
aimodelpoll.iloc[:, 2:] = aimodelpoll.iloc[:, 2:].round(3)
```

```python
aimodelpoll.head(5)
```

```python
aimodelpoll.info()
```

```python
original_file = 'aimodelpoll.csv'
aimodelpoll.to_csv(original_file, index=False)
```

```python
llmach.info()
```

```python
for col in ['Result', 'Human result']:
    llmach[col] = (
        llmach[col]
        .astype(str)                                 
        .str.replace(r'[\s%,]', '', regex=True)      
        .str.replace(',', '.', regex=False)         
        .replace(['', 'nan', 'NaN', '-', '--'], np.nan)
        .pipe(pd.to_numeric, errors='coerce')        
    )
llmach['Result'] = llmach['Result'].round(2)
llmach['Human result'] = llmach['Human result'].round(2)
```

```python
llmach.head(5)
```

```python
llmach = llmach.drop(columns=llmach.columns[8])
llmach = llmach.drop(columns=llmach.columns[7])
```

```python
llmach.info()
```

```python
original_file = 'llmachievements.csv'
llmach.to_csv(original_file, index=False)
```

```python
keep_columns = [
    'Model',
    'Domain',
    'Task',
    'Organization',
    'Publication date',
    'Link',
    'Parameters',
    'Training compute (FLOP)',
    'Training time (hours)',
    'Training hardware',
    'Model accessibility',
    'Country (of organization)',
    'Organization categorization',
    'Training code accessibility',
    'Hugging Face developer id'
]

rename_map = {
    'Model': 'model',
    'Domain': 'domain',
    'Task': 'tasks',
    'Organization': 'organization',
    'Publication date': 'release_date',
    'Link': 'reference_url',
    'Parameters': 'parameters',
    'Training compute (FLOP)': 'training_flop',
    'Training time (hours)': 'training_hours',
    'Training hardware': 'training_hardware',
    'Model accessibility': 'accessibility',
    'Country (of organization)': 'country',
    'Organization categorization': 'org_type',
    'Training code accessibility': 'code_available',
    'Hugging Face developer id': 'hf_repo_id'
}

df_mops = modelview[keep_columns].rename(columns=rename_map).copy()
df_mops['release_date'] = pd.to_datetime(df_mops['release_date'], errors='coerce')
df_mops = df_mops.sort_values('release_date', ascending=False)
print(df_mops.shape)
print(df_mops.head())
```

```python
original_file = 'overviewmodels.csv'
df_mops.to_csv(original_file, index=False)
```
