```python
# create athena db for query 
```

```python
import boto3
import sagemaker

sess = sagemaker.Session()
bucket = sess.default_bucket()
role = sagemaker.get_execution_role()
region = boto3.Session().region_name
```

```python
from pyathena import connect
```

```python
database_name = "dsoaws"
```

```python
s3_staging_dir = "s3://{0}/athena/staging".format(bucket)
```

```python
conn = connect(region_name=region, s3_staging_dir=s3_staging_dir)
```

```python
statement = "CREATE DATABASE IF NOT EXISTS {}".format(database_name)
print(statement)
```

```python
import pandas as pd
pd.read_sql(statement, conn)
```

```python
statement = "SHOW DATABASES"
df_show = pd.read_sql(statement, conn)
df_show.head(5)
```

```python
database_name = "dsoaws"
llmtable_name_csv = "llmtable"
s3_public_path_csv = "s3://sagemaker-us-east-1-883727203134/table".format(bucket)
```

```python
llmexpo = pd.read_csv('Open_LLM-Perf_Leaderboard.csv')
llmexpo.head()
```

```python
statement = """CREATE EXTERNAL TABLE IF NOT EXISTS {}.{}(
         Model string,
         Type string,
         Class string,
         Backend string,
         Dtype string,
         Optimizations string,
         Throughput_tokens float,
         peak_memory_mb int,
         score int,
) ROW FORMAT DELIMITED FIELDS TERMINATED BY '\\t' LINES TERMINATED BY '\\n' LOCATION '{}'
TBLPROPERTIES ('compressionType'='gzip', 'skip.header.line.count'='1')""".format(
    database_name, llmtable_name_csv, s3_public_path_csv
)
print(statement)
```
